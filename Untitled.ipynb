```python
d1=pd.read_csv("student-mat.csv",sep=";")
d2=pd.read_csv("student-por.csv",sep=";")
merge_columns=["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"]
d1_selected=d1[merge_columns]
d2_selected=d2[merge_columns]
d3=pd.merge(d1_selected,d2_selected,on=merge_columns,how="inner")
d3.to_csv("student_merged.csv",index=False)
```

```python
# Parental education level (Average of mother’s and father’s education)
d3["ParentalEdu"] = (d3["Medu"] + d3["Fedu"]) / 2
d3["ParentalEdu"] 
```

```python
plt.figure(figsize=(8, 5))
sn.countplot(x=d3["ParentalEdu"], palette="coolwarm")
plt.xlabel("Parental Education Level")
plt.ylabel("Number of Students")
plt.title("Parental Education Level Count")
plt.show()
```


## 1. Level 1 → Primary Education (Elementary School)

## 2. Level 2 → Lower Secondary Education (Middle School)

## 3. Level 3 → Upper Secondary Education (High School)

## 4. Level 4 → Higher Education (University/College)

# most common education level is 4.0
# least common education level 0.5

```python
d3["U"]=(d3["address"].str.strip()=="U").astype(int)
d3["U"]
```

```python
urban_count=d3["U"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(urban_count, labels=["Urban", "Rural"], autopct="%1.1f%%", colors=["red", "blue"], startangle=90)
plt.title("Urban vs. Rural Student Proportion")
plt.show()
```
